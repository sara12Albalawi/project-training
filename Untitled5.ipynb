```python
# الخطوة 1: تثبيت المكتبات الضرورية
# نقوم بتثبيت المكتبات التي نحتاجها لتحليل البيانات والرسم البياني.
!pip install yfinance
!pip install pandas
!pip install plotly

# الخطوة 2: استيراد المكتبات
# نقوم باستيراد المكتبات التي سنستخدمها في الكود.
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# الخطوة 3: تحديد السهم والبيانات
# هنا يمكنك تغيير رمز السهم (Ticker) والفترة الزمنية التي تريد تحليلها.
ticker_symbol = "2222.SR"  # رمز سهم أرامكو السعودية
start_date = "2024-01-01"
end_date = "2024-09-06"

# الخطوة 4: تحميل بيانات السهم
# نستخدم مكتبة yfinance لتحميل البيانات التاريخية للسهم.
data = yf.download(ticker_symbol, start=start_date, end=end_date)

# الخطوة 5: حساب المؤشرات الفنية
# سنقوم بحساب المتوسط المتحرك البسيط (SMA) ومؤشر القوة النسبية (RSI).
data['SMA_50'] = data['Close'].rolling(window=50).mean()
data['SMA_200'] = data['Close'].rolling(window=200).mean()

# حساب مؤشر القوة النسبية (RSI)
delta = data['Close'].diff()
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()
rs = avg_gain / avg_loss
data['RSI'] = 100 - (100 / (1 + rs))

# الخطوة 6: إنشاء الرسم البياني
# نستخدم مكتبة Plotly لإنشاء رسم بياني تفاعلي.
# نقوم بإنشاء رسم بياني مع لوحتين: واحدة للسعر والأخرى لمؤشر RSI.
fig = make_subplots(rows=2, cols=1, shared_xaxes=True,
                    vertical_spacing=0.1,
                    row_heights=[0.7, 0.3],
                    subplot_titles=[f'تحليل السهم لـ {ticker_symbol}', 'مؤشر القوة النسبية (RSI)'])

# إضافة مخطط الشموع (Candlestick chart) للسعر
fig.add_trace(go.Candlestick(x=data.index,
                             open=data['Open'],
                             high=data['High'],
                             low=data['Low'],
                             close=data['Close'],
                             name='سعر السهم'),
              row=1, col=1)

# إضافة المتوسطات المتحركة
fig.add_trace(go.Scatter(x=data.index, y=data['SMA_50'], mode='lines',
                         name='متوسط 50 يوم', line=dict(color='blue', width=1)),
              row=1, col=1)
fig.add_trace(go.Scatter(x=data.index, y=data['SMA_200'], mode='lines',
                         name='متوسط 200 يوم', line=dict(color='orange', width=1)),
              row=1, col=1)

# إضافة مؤشر RSI
fig.add_trace(go.Scatter(x=data.index, y=data['RSI'], mode='lines',
                         name='RSI', line=dict(color='purple', width=1)),
              row=2, col=1)

# إضافة خطوط مؤشر RSI لمناطق التشبع (30) والبيع المفرط (70)
fig.add_hline(y=70, line_dash="dash", line_color="red", row=2, col=1)
fig.add_hline(y=30, line_dash="dash", line_color="green", row=2, col=1)

# تحديث تخطيط الرسم البياني
fig.update_layout(xaxis_rangeslider_visible=False,
                  title_text=f"تحليل فني لأسهم {ticker_symbol}",
                  height=700,
                  template="plotly_dark")

# عرض الرسم البياني
fig.show()
```
